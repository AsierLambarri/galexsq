```python
import yt
import pandas as pd
import numpy as np
from astropy.table import Table
from unyt import unyt_quantity, unyt_array
import smplotlib
from tqdm import tqdm

def load_ftable(fn):
    """Loads astropy tables formated with ascii.fixed_width and sep='\t'. These tables are human readable but
    a bit anoying to read with astropy because of the necessary keyword arguments. This functions solves that.
    Useful for tables that need to be accessed in scripts but be readable (e.g. csv are NOT human readable).

    Equivalent to : Table.read(fn, format="ascii.fixed_width", delimiter='\t')

    Parameters
    ----------
    fn : string, required
        Path to Formated Table file.

    Returns
    -------
    ftable : pd.DataFrame
    """
    return Table.read(fn, format="ascii.fixed_width", delimiter="\t").to_pandas()
```

```python
CompleteTree = pd.read_csv("../merger_trees_csv/GEAR_CompleteTree.csv")
HostTree = pd.read_csv("../merger_trees_csv/GEAR_HostTree.csv")
```

```python
candidates = pd.read_csv("/home/asier/StructuralAnalysis/satellite_tables/GEAR_satellitesV5_Rvir5.0.csv")

fcand = pd.read_csv("/home/asier/StructuralAnalysis/satellite_tables/GEAR_candidatesV5.csv")
```

```python
candidates
```

```python
snapnum = 756

width = (1000, 'kpccm')
```

```python
filtered_halos  = CompleteTree[(CompleteTree['Snapshot'] == snapnum) &
                                (CompleteTree['mass'] > 1E7)]
```

```python
host = HostTree[HostTree['Snapshot'] == snapnum]
host_subid = host['Sub_tree_id'].values
```

```python
snapEquiv = load_ftable("./GEAR_equivalenceV2.dat")
file = snapEquiv[snapEquiv['snapshot'] == snapnum]['snapname'].values[0]
pdir = "/media/asier/T7/GEAR/"
fn = pdir + file
fn
```

```python
startype = ("PartType1", 'Masses')
dmtype = ("PartType2", "Masses")
gastype = ("PartType0", "Density")
```

```python
ds = yt.load(fn) 
ds.code_length = ds.length_unit
ad = ds.all_data()
z = ds.current_redshift

center = host[['position_x', 'position_y', 'position_z']].values[0].astype(float) * ds.units.kpccm

print(f'The center will be located at {center.in_units("kpccm")}')

fig = plt.figure(figsize=(30, 30))
grid = AxesGrid(
    fig, (-1.8, -1.8, 1.8, 1.8),
    nrows_ncols=(2, 2),
    axes_pad=1.1,
    cbar_mode='each',  # Colorbar mode: 'each' for individual colorbars
    cbar_location='right',  # Colorbar position for each subplot
    cbar_pad=0,  # Padding between plot and colorbar
    share_all=False
)


# Gas
p = yt.ProjectionPlot(ds,'z',gastype, center=center, width=width)
p.set_unit(gastype, "Msun/kpc**2")
p.set_font(
     { "size": 20,
    }

)
for i in tqdm(range(0, len(filtered_halos))):
    sub_tree_id = filtered_halos['Sub_tree_id'].iloc[i]
    position_x = -(filtered_halos['position_x'].iloc[i] - (1+z) * center[0].in_units("kpc").value - 500)/1000 
    position_y = -(filtered_halos['position_y'].iloc[i] - (1+z) * center[1].in_units("kpc").value - 500)/1000 
    position_z = -(filtered_halos['position_z'].iloc[i] - (1+z) * center[2].in_units("kpc").value - 500)/1000 
    virial_radius = filtered_halos['virial_radius'].iloc[i] 

    if sub_tree_id == 789167:
        p.annotate_sphere([1-position_x, 1-position_y], radius=(1*virial_radius, 'kpccm'),coord_system='axis',
                    circle_args={"color": "green", "linestyle": "dashed", "linewidth": 2})


p.set_axes_unit('kpccm')
p.annotate_timestamp(redshift=True, text_args={"color":"black"})
p.set_xlabel('x [comoving kpc]')
p.set_ylabel('y [comoving kpc]')

plot = p.plots[gastype]
plot.figure = fig
plot.axes = grid[0].axes
plot.cax = grid.cbar_axes[0]

p.render()


# DM
p = yt.ParticleProjectionPlot(ds,'z',dmtype, center=center, width=width, density=True)
p.set_unit(dmtype, "Msun/kpc**2")
p.set_font(
     {  
        "size": 20,
    }

)
p.set_axes_unit('kpccm')
p.annotate_timestamp(redshift=True, text_args={"color":"black"})
p.set_xlabel('x [comoving kpc]')
p.set_ylabel('')
for i in tqdm(range(0, len(filtered_halos))):
    sub_tree_id = filtered_halos['Sub_tree_id'].iloc[i]
    position_x = -(filtered_halos['position_x'].iloc[i] - (1+z) * center[0].in_units("kpc").value - 500)/1000 
    position_y = -(filtered_halos['position_y'].iloc[i] - (1+z) * center[1].in_units("kpc").value - 500)/1000 
    position_z = -(filtered_halos['position_z'].iloc[i] - (1+z) * center[2].in_units("kpc").value - 500)/1000 
    virial_radius = filtered_halos['virial_radius'].iloc[i] 

    if sub_tree_id == 1:
        p.annotate_sphere([1-position_x, 1-position_y], radius=(1*virial_radius, 'kpccm'),coord_system='axis',
                    circle_args={"color": "green", "linestyle": "dashed", "linewidth": 2})
    else:    
        p.annotate_sphere([1-position_x, 1-position_y],
                        radius=(0.5*virial_radius, 'kpccm'),
                         coord_system='axis', circle_args={"color": "white", "linewidth": 2})




plot = p.plots[dmtype]
plot.figure = fig
plot.axes = grid[1].axes
plot.cax = grid.cbar_axes[1]
p.render()



# Stars
p = yt.ParticleProjectionPlot(ds,'z',startype, center=center, width=width, density=True, deposition="ngp")
p.set_unit(startype, "Msun/kpc**2")
p.set_font(
     {  
        "size": 20,
    }

)
for i in tqdm(range(0, len(filtered_halos))):
    m = filtered_halos['mass'].iloc[i]
    sub_tree_id = filtered_halos['Sub_tree_id'].iloc[i]
    posx = filtered_halos['position_x'].iloc[i] - (1+z) * center[0].in_units("kpc").value
    posy = filtered_halos['position_y'].iloc[i] - (1+z) * center[1].in_units("kpc").value
    posz = filtered_halos['position_z'].iloc[i] - (1+z) * center[2].in_units("kpc").value
    position_x = -(posx - 500)/1000 
    position_y = -(posy - 500)/1000 
    position_z = -(posz - 500)/1000 
    virial_radius = filtered_halos['virial_radius'].iloc[i] 

    if (sub_tree_id in fcand['Sub_tree_id'].values):
        if (fcand[fcand['Sub_tree_id'] == sub_tree_id]['stellar_mass'].values>0):
            p.annotate_sphere([1-position_x, 1-position_y], radius=(1*virial_radius, 'kpccm'),coord_system='axis',
                        circle_args={"color": "red", "linestyle": "solid", "linewidth": 2})
            if np.abs(posx) < width[0]/2 and np.abs(posy) < width[0]/2:
                p.annotate_text([1-position_x, 1-position_y-virial_radius/2/width[0]], f"{int(sub_tree_id)}", coord_system="axis",
                            text_args={"color": "red", "fontsize":10}
                           )

    if sub_tree_id == host_subid :
        p.annotate_sphere([1-position_x, 1-position_y], radius=(1.5*virial_radius, 'kpccm'),coord_system='axis',
                    circle_args={"color": "green", "linestyle": "dashed", "linewidth": 2})

    if sub_tree_id in candidates['Sub_tree_id'].values:
        p.annotate_sphere([1-position_x, 1-position_y], radius=(1*virial_radius, 'kpccm'),coord_system='axis',
                    circle_args={"color": "green", "linestyle": "dashed", "linewidth": 2})
        if np.abs(posx) < width[0]/2 and np.abs(posy) < width[0]/2:
            p.annotate_text([1-position_x, 1-position_y-virial_radius/2/width[0]], f"{int(sub_tree_id)}", coord_system="axis",
                        text_args={"color": "green", "fontsize":10}
                       )
            
    else:
        if m > 1E9:
            p.annotate_sphere([1-position_x, 1-position_y], radius=(1*virial_radius, 'kpccm'),coord_system='axis',
                    circle_args={"color": "blue", "linestyle": "dashed", "linewidth": 2})  
            if np.abs(posx) < width[0]/2 and np.abs(posy) < width[0]/2:
                p.annotate_text([1-position_x, 1-position_y-virial_radius/2/width[0]], f"{int(sub_tree_id)}", coord_system="axis",
                        text_args={"color": "blue", 'ha':"center","fontsize":10}
                       )

p.set_axes_unit('kpccm')
p.annotate_timestamp(redshift=True, text_args={"color":"black"})
#p.set_xlabel('z [comoving kpc]')
#p.set_ylabel('y [comoving kpc]')

plot = p.plots[startype]
plot.figure = fig
plot.axes = grid[2].axes
plot.cax = grid.cbar_axes[2]

p.render()

grid[3].set_axis_off()
grid.cbar_axes[3].set_visible(False)


plt.show()
```
