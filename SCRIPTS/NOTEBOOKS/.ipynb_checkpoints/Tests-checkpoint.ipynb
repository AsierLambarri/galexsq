```python
def _collapse_to_longest_true(arr):
    """
    Given a boolean array, returns a new array where only the longest contiguous block 
    of True values is preserved, and all other True values are set to False.
    
    This function pads the array and uses np.diff to locate transitions.
    """
    arr = np.asarray(arr, dtype=bool)
    
    # If there are no True values, just return a copy.
    if not arr.any():
        return arr.copy()
    
    # Pad the array with False at both ends.
    padded = np.r_[False, arr, False]
    
    # Compute the differences between consecutive elements.
    # When converting booleans to int: False->0, True->1.
    d = np.diff(padded.astype(int))
    
    # A diff of 1 indicates a transition from False to True: these are our start indices.
    starts = np.where(d == 1)[0]
    
    # A diff of -1 indicates a transition from True to False: these are our end indices.
    # Since we padded on the left, subtract 1 to map back to original array indices.
    ends = np.where(d == -1)[0] - 1
    
    # Compute the lengths of each contiguous True block.
    lengths = ends - starts + 1
    
    # Identify the block with the maximum length.
    max_idx = np.argmax(lengths)
    longest_start = starts[max_idx]
    longest_end = ends[max_idx]
    
    # Build a new array: set only the indices between longest_start and longest_end to True.
    new_arr = np.zeros_like(arr, dtype=bool)
    new_arr[longest_start:longest_end+1] = True
    return new_arr
```

```python
import numpy as np
from sklearn.cluster import KMeans
import kneed

def detect_flat_regions(x, y):
    """
    Identify flat portions of a function defined by x and y arrays.
    Optimized for NumPy and handles NaNs by processing only contiguous
    segments of valid (finite) data. Ensures that detected flat regions
    have a minimum length of at least 10% of the sample size.

    Parameters:
      x (array-like): 1D array of x values.
      y (array-like): 1D array of y values.
    
    Returns:
      flat_regions (list of tuples): Each tuple is (x_start, x_end) of a valid flat region.
      slopes (np.array): Computed slope at each point (NaN where not computed).
      flat_mask (np.array): Boolean mask where True indicates a flat point.
    """
    x = np.asarray(x, dtype=float)
    y = np.asarray(y, dtype=float)
    n = len(x)

    if n < 2:
        return [], np.full(n, np.nan), np.full(n, False)

    # Create a mask for valid (finite) data points
    valid_mask = np.isfinite(x) & np.isfinite(y)
    
    # Initialize slopes as NaN
    slopes = np.full(n, np.nan, dtype=float)
    
    # Identify indices of valid data
    valid_idx = np.flatnonzero(valid_mask)
    if valid_idx.size < 2:
        return [], slopes, np.full(n, False)
    
    # Split valid indices into contiguous segments
    breaks = np.where(np.diff(valid_idx) != 1)[0] + 1
    segments = np.split(valid_idx, breaks)
    
    # Compute slopes for each contiguous segment
    for seg in segments:
        if seg.size < 2:
            continue  # Need at least two points to compute a derivative
        idx = seg
        # Compute forward difference for the first point
        slopes[idx[0]] = (y[idx[1]] - y[idx[0]]) / (x[idx[1]] - x[idx[0]])
        # Compute backward difference for the last point
        slopes[idx[-1]] = (y[idx[-1]] - y[idx[-2]]) / (x[idx[-1]] - x[idx[-2]])
        # Compute central differences for interior points
        if idx.size > 2:
            i_inner = idx[1:-1]
            slopes[i_inner] = (y[idx[2:]] - y[idx[:-2]]) / (x[idx[2:]] - x[idx[:-2]])

    # Only consider slopes computed at valid indices
    valid_slopes = slopes[valid_mask]
    if valid_slopes.size < 2:
        return [], slopes, np.full(n, False)

    # Reshape the absolute slopes for clustering
    abs_slopes = np.abs(valid_slopes).reshape(-1, 1)

    # Use KMeans with k=2 to classify slopes into "flat" and "non-flat"
    kmeans = KMeans(n_clusters=4, random_state=0, n_init='auto').fit(abs_slopes)
    labels = kmeans.labels_
    centers = kmeans.cluster_centers_

    # The cluster with the smaller center is assumed to be the "flat" cluster
    flat_cluster = np.argmin(centers)

    # Build a full boolean mask for all points: only valid points are classified
    flat_mask = np.full(n, False, dtype=bool)
    flat_mask[valid_mask] = (labels == flat_cluster)

    # Group contiguous indices in the full array that are marked as flat
    flat_indices = np.flatnonzero(flat_mask)
    min_length = max(3, int(0.1 * n))  # 10% of total sample size (at least 1 point)

    flat_regions = []
    if flat_indices.size > 0:
        # Find breaks in the flat indices
        seg_breaks = np.where(np.diff(flat_indices) != 1)[0] + 1
        flat_segments = np.split(flat_indices, seg_breaks)
        for seg in flat_segments:
            if len(seg) >= min_length:
                flat_regions.append((x[seg[0]], x[seg[-1]]))  # Store (x_start, x_end)

    return flat_regions, slopes, flat_mask

# Example usage:
if __name__ == '__main__':
    
    mask = surf_mag_profile > 1
    flat_regions, slopes, flat_mask = detect_flat_regions(radii[mask], surf_mag_profile[mask])
    print("Flat regions (by x interval):", flat_regions)


kneedle_convex = kneed.KneeLocator(radii[mask], surf_mag_profile[mask], S=1, curve="convex", direction="increasing", online=True)
kneedle_concave =  kneed.KneeLocator(radii[mask], surf_mag_profile[mask], S=1, curve="concave", direction="increasing", online=True)

min_concave = min(kneedle_concave.all_knees)
min_convex = min(kneedle_convex.all_knees)
print(kneedle_concave.all_knees)
print(kneedle_convex.all_knees)
```

```python
def plummer(r, M, a):
    """Plummer model
    """ 
    return 3 * M / ( 4 * np.pi * a ** 3 ) * ( 1 + (r/a)**2 ) ** (-5/2)
    
def surface_plummer(R, M, a):
    """Surface density plummer
    """ 
    return ( M * a ** 2 ) / np.pi * 1 / ( a**2 + R**2 ) ** 2

def veldisp_plummer(r, M, a):
    return 4.300917270038e-06 * M / 6 * 1 / np.sqrt(r**2 + a**2)
```
