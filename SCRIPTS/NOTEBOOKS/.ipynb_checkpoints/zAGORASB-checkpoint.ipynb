```python
import yt
import numpy as np
import pandas as pd
from scipy.spatial import KDTree

from pNbody import *
from astropy import units as u

from modules.loaders import ARTI_loader, GEAR_loader
```

```python
ds = ARTI_loader("/media/asier/EXTERNAL_USBA/Cosmo_v18/10MpcBox_csf512_04078.d")
```

```python
host = pd.read_csv("../merger_trees_csv/ART_HostTree.csv")
host
```

```python
center = ds.arr( host[["position_x", "position_y", "position_z"]].iloc[0].values, 'kpccm')
center
```

```python
def _compute_hsml(field, data):
    tree = KDTree(data["stars", "particle_position"].to('kpc').value )
    distances, _ = tree.query(data["stars", "particle_position"].to('kpc').value , k=33) 

    hsml = distances[:, -1] * data["stars", "particle_position"].units
    return hsml

ds.add_field(
    ("stars", "hsml"),
    function=_compute_hsml,
    sampling_type="local",
    units='kpc',
    force_override=True
)
```

```python
sp_vir = ds.sphere(center, (400, 'kpc'))
```

```python
pos = sp_vir["stars", "particle_position"].to('kpc').value 
vel = sp_vir["stars", "particle_velocity"].to('km/s').value 

mass = ds.arr(sp_vir["stars", "initial_mass"], "code_mass").to('1E10 * Msun').value
mh = sp_vir["stars", "metallicity"].value
age = sp_vir["stars", "age"].to('Gyr').value
hsml = sp_vir["stars", "hsml"].to("kpc").value
```

```python
outputfile = "art_0845.hdf5"

nb = Nbody(status='new', pos=pos, mass=mass, vel=vel, rsp=hsml, ftype='swift')
nb.set_tpe(4)
nb.age                =  age
nb.mh                 =  mh

u_Length   = 1 * u.kpc
u_Mass     = 1E10 * u.M_sun
u_Velocity = 1 * u.km/u.s
u_Time     = u_Length/u_Velocity 
toMsol     = u_Mass.to(u.M_sun).value

nb.UnitLength_in_cm         = u_Length.to(u.cm).value
nb.UnitMass_in_g            = u_Mass.to(u.g).value        
nb.UnitVelocity_in_cm_per_s = u_Velocity.to(u.cm/u.s).value
nb.Unit_time_in_cgs         = u_Time.to(u.s).value

nb.hubblefactorcorrection      = False
nb.comovingtoproperconversion  = False
nb.atime                       = 1
nb.cmcenter()
nb.cvcenter()
nb.rename(outputfile)
```
