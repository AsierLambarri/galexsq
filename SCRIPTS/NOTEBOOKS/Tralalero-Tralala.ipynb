```python
import yt
import numpy as np
import pandas as pd
from galexquared.class_methods import load_ftable
```

```python
table = load_ftable("GEAR_equivalence.dat"); table
```

```python
merger = pd.read_csv("merger_trees_csv/fullGEAR_HostTree.csv")
```

```python
pdir = "/media/asier/T7/GEAR/"
table["redshift_RCT"] = pd.Series()
table["time"] = pd.Series()

for index, line in table.iterrows():
    try:
        ds = yt.load(pdir + line["snapname"])
        time = ds.current_time.to("Gyr").value
    except:
        print(f"Failed with {line['snapname']}")
        time = np.nan

    z_rct = merger[merger["Snapshot"] == line["snapshot"]]["Redshift"].values[0]

    table.loc[index, "redshift_RCT"] = z_rct
    table.loc[index, "time"] = time
```
